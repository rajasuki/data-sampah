```python
import pandas as pd

df = pd.read_csv("sampah.csv")

df = df[['nama_kabupaten_kota', 'jumlah_sampah', 'tahun']]

df.columns = ['Kabupaten_Kota', 'Produksi_Sampah_Ton', 'Tahun']

print(df.head())
```

```python
total_per_tahun = {}

for index, row in df.iterrows():
    tahun = row['Tahun']
    produksi = row['Produksi_Sampah_Ton']

    if tahun not in total_per_tahun:
        total_per_tahun[tahun] = 0

    total_per_tahun[tahun] += produksi

df_total_tahun = pd.DataFrame(
    total_per_tahun.items(),
    columns=['Tahun', 'Total_Produksi_Sampah_Ton']
)

print(df_total_tahun)
```

```python
total_kab_tahun = {}

for index, row in df.iterrows():
    key = (row['Kabupaten_Kota'], row['Tahun'])
    produksi = row['Produksi_Sampah_Ton']

    if key not in total_kab_tahun:
        total_kab_tahun[key] = 0

    total_kab_tahun[key] += produksi

df_kab_tahun = pd.DataFrame([
    {
        'Kabupaten_Kota': k[0],
        'Tahun': k[1],
        'Total_Produksi_Sampah_Ton': v
    }
    for k, v in total_kab_tahun.items()
])
print(df_kab_tahun)

```

```python
with pd.ExcelWriter("hasil_analisis_sampah.xlsx") as writer:
    df_total_tahun.to_excel(writer, sheet_name="Total_Per_Tahun", index=False)
    df_kab_tahun.to_excel(writer, sheet_name="Per_Kabupaten_Tahun", index=False)
```
